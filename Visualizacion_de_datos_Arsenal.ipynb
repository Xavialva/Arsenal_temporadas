```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=('/content/drive/MyDrive/PTDataScience/VisualDatos/Week51.xlsx')
```

```python
df= pd.read_excel(df)
```

```python
df
```

```python
table_position=df[['Season','Position']]
table_position
```

```python
plt.figure(figsize=(10,6))
plt.bar(table_position['Season'],table_position['Position'])
plt.xticks(rotation=90)
plt.title('Tabla de posiciones del Arsenal en las últimas 29 temporadas (Mitad de temporada)',fontsize= 15)


```

# Podemos observar que el inicio de temporada '2020/21' es el peor inicio de temporada de las últimas 29 temporadas(mitad)


```python
points=df[['Season','Points']]
```

```python
points
```

```python
plt.figure(figsize=(10,6))
plt.bar(points['Season'],points['Points'])
plt.ylim(top=45)
plt.xticks(rotation=90)
plt.title('Puntos por Mitad de Temporada',fontsize=15)
```

# Podemos observar que las dos últimas temporadas son las que menos punto hizo al inicio de temporada

```python
goals_scored=df[['Season','Goals For']]  
```

```python
goals_scored
```

```python
nums_goals= goals_scored['Goals For'].values
```

```python
nums_goals
```

```python
season_goals= goals_scored['Season'].values
```

```python
season_goals
```

```python
plt.figure(figsize=(10,6))
plt.bar(season_goals,nums_goals)
plt.ylim(top=45)
plt.xticks(rotation=90)
plt.title('Número de goles por mitada de temporada',fontsize=15)
```

# La temporada '2004/05' y '2009/10' fueron un 'éxito' en goles (la última temporada no nos sorprende).

```python
goals_against=df[['Season','Goals Against']]
```
